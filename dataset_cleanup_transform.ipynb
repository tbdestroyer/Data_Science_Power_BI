```python
import pandas as pd
#import the CSV data file into a dataframe
crash_df = pd.read_csv('Maryland_Statewide_Vehicle_Crashes.csv', low_memory=False)
```

```python
crash_df.head()
```

```python
crash_df.columns
```

```python
#drop some unneeded columns
crash_df.drop(columns=["MUNI_DESC","MUNI_CODE", "JUNCTION_DESC", 'JUNCTION_CODE','LANE_CODE',
                       'RD_COND_DESC', 'RD_COND_CODE','AREA_CODE', 'RTE_SUFFIX', 'LOG_MILE',
                       'LOGMILE_DIR_FLAG_DESC','LOGMILE_DIR_FLAG','DISTANCE',
                       'FEET_MILES_FLAG_DESC','FEET_MILES_FLAG', 'DISTANCE_DIR_FLAG', 'REFERENCE_NO',
                       'REFERENCE_TYPE_CODE', 'REFERENCE_SUFFIX', 'REFERENCE_ROAD_NAME','LOCATION'], inplace=True)
```

```python
#Convert county names to upper case
crash_df['COUNTY_DESC'] = crash_df['COUNTY_DESC'].str.upper()
```

```python
crash_df.head()
```

```python
#Select only data after year 2017
crash_df = crash_df[crash_df['YEAR'] > 2017]
```

```python
#check the data frame
crash_df.shape
```

```python
#drop rows with not available County description
crash_df =crash_df.dropna(subset=['COUNTY_DESC'])
```

```python
crash_df.shape
```

```python
crash_df.to_csv('Maryland_state_crash.csv')
```

```python
#Read the registartion dataset
origreg_df = pd.read_csv('MVA_VEHICLE_REGISTRATION_by_COUNTY_from_2010_to_2022.csv')
```

```python
origreg_df.head() #check the original dataset provided
```

```python

origreg_df.columns
```

```python
#Transform the vehicke registrations into a csv file with 'year', 'county', 'registration' columns 
county=[]
year=[]
registrations=[]

for col in origreg_df.columns:
    for idx in origreg_df.index:
        if idx > 7 and col != 'Fiscal Year':
            year.append(idx+2010)
            county.append(col)
            registrations.append(origreg_df.at[idx, col])
```

```python
reg_df= pd.DataFrame({'year':year, 
         'county':county,
         'registrations':registrations})
reg_df
```

```python
#Fix county names between two tables to match
reg_df.loc[reg_df["county"] == "ST MARY'S", "county"] = "ST. MARY'S"  # has a dot
reg_df.loc[reg_df["county"] == "TALBOT ", "county"] = "TALBOT" # has a space
reg_df.head(120)
```

```python
# Iterate the registration dataframe and match the county and year to the crash dataframe
#and select the rows (count of accident entries) and append to the accident list
accidents=[]
for idx in range(len(reg_df)):
    accidents.append(len(crash_df[(crash_df['COUNTY_DESC'] == reg_df.at[idx, 'county']) & (crash_df['YEAR'] == reg_df.at[idx, 'year'])]))

len(accidents)
```

```python
reg_df= pd.DataFrame({'year':year, 
         'county':county,
         'registrations':registrations,
         'accidents': accidents})
reg_df
```

```python
#write the transformed dataframe into a cvs file
reg_df.to_csv('Maryland_state_registrations.csv',index = False)
```
